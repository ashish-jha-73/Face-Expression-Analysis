```python
import cv2
import numpy as np
import os
import random
from tensorflow.keras.models import load_model

model = load_model("fea-iter-2.keras")

emotion_labels = [
    "angry",
    "disgust",
    "fear",
    "happy",
    "neutral",
    "sad",
    "surprise"
]

test_root = "../archive/test"   

def preprocess_image(img):
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    resized = cv2.resize(img_rgb, (48, 48))
    img_float = resized.astype("float32")  
    final_img = np.expand_dims(img_float, axis=0) 
    return final_img


random_class = random.choice(emotion_labels)
class_path = os.path.join(test_root, random_class)

valid_ext = (".jpg", ".jpeg", ".png")
images = [f for f in os.listdir(class_path) if f.lower().endswith(valid_ext)]

random_image = random.choice(images)
image_path = os.path.join(class_path, random_image)

print(f"\nRandom Class: {random_class}")
print(f"Random Image: {random_image}")


img = cv2.imread(image_path)
# cv2.imshow("Random Test Image", img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()

processed = preprocess_image(img)
preds = model.predict(processed)
pred_label = emotion_labels[np.argmax(preds)]

print(f"\nTrue Label: {random_class}")
print(f"Predicted Label: {pred_label}")
```
